```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
# Make sure to replace 'zomato.csv' with your dataset filename
data = pd.read_csv("/content/Dataset .csv")

# Display first few rows
print("Dataset preview:")
print(data.head())

# --- 1️⃣ Most Common Price Range ---
most_common_price_range = data['Price range'].mode()[0]
print(f"\nMost Common Price Range: {most_common_price_range}")

# --- 2️⃣ Average Rating for Each Price Range ---
avg_rating_by_price = data.groupby('Price range')['Aggregate rating'].mean().reset_index()
print("\nAverage Rating for Each Price Range:")
print(avg_rating_by_price)

# --- 3️⃣ Identify the Color with Highest Average Rating ---
color_rating = data.groupby('Rating color')['Aggregate rating'].mean().reset_index()
highest_color = color_rating.loc[color_rating['Aggregate rating'].idxmax()]
print(f"\nColor with Highest Average Rating: {highest_color['Rating color']} "
      f"({highest_color['Aggregate rating']:.2f})")

# --- Optional Visualization ---
plt.figure(figsize=(8,5))
sns.barplot(x='Price range', y='Aggregate rating', data=avg_rating_by_price, palette='viridis')
plt.title('Average Rating by Price Range')
plt.xlabel('Price Range')
plt.ylabel('Average Rating')
plt.show()

plt.figure(figsize=(8,5))
sns.barplot(x='Rating color', y='Aggregate rating', data=color_rating, palette='plasma')
plt.title('Average Rating by Rating Color')
plt.xlabel('Rating Color')
plt.ylabel('Average Rating')
plt.show()
```
